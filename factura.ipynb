```python
from datetime import datetime
from mimetypes import init
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas
from IPython.display import Image




equipos={'pc APPLE':6000000,'pc Asus':4500000,'pc DELL':3000000,'pc Huawei':4600000,'pc Lenovo':5000000,'Huawei':2000000,'Apple':9300000,'Xiaomi':3200000,'LG':1200000,'Motorola':1500000}



class factura:

    date = datetime.now()
    fecha=date.date()
    hora=date.time()
    
    
    totalpagar=0
    
  
    separacion="========================================="

   

    def __init__(self,nombre,identificacion):
        self.nombre=nombre
        self.identificacion=identificacion

    

    
    def factura_com(self):
        w, h = letter
        c=canvas.Canvas('recibo caja1.pdf',pagesize=letter)
        c.setFont('Helvetica',10)
        c.rect(200,h-410,280,350)

        print('               TECNOSTORE','\n            TEL: 3003232112','\n                FRENESI','\n              CRA 1 #3-08')
        c.drawString(210,704,'                              TECNOSTORE')
        c.drawString(210,694,'                            TEL: 30032321112')
        c.drawString(210,684,'                                    NEIVA')
        c.drawString(210,674,'                                CRA 1 #3-08')

        print('FECHA:',self.fecha,' HORA:',self.hora,'\nJUAN DE LA CRUZ GERRA')
        print(self.separacion)
        c.drawString(210,660,f"  FECHA:{self.fecha}   HORA: {self.hora}")
        c.drawString(210,650,'  CAJERO: JUAN DE LA CRUS GERRA')
        c.drawString(210,640,self.separacion)

        print('CLIENTE:',self.nombre,'\nIDENTIFICACION:',self.identificacion)
        print(self.separacion)
        c.drawString(210,630,f'  CLIENTE: {self.nombre}')
        c.drawString(210,620,f'  IDENTIFICACION: {self.identificacion}')
        c.drawString(210,610,self.separacion)
        
        print('Uds DESCRIPCION      PRECIO      TOTAL')
        print(self.separacion)
        c.drawString(210,600,'  Uds  DESCRIPCION      PRECIO      TOTAL')
        c.drawString(210,590,self.separacion)


        y=580
        while True:
            self.articulo=str(input('nombre del articulo'))
            self.precio=equipos[self.articulo]
            self.unidades=int(input('unidades'))
            self.total=self.precio*self.unidades
            print(self.unidades,' ',self.articulo,'      ',self.precio,'     ',self.total)
            self.totalpagar=self.totalpagar+self.total
            opcion=str(input('MAS PRODUCTOS(MP) - TERMINAR(TR)'))
            c.drawString(210,y,f'     {self.unidades}     {self.articulo}              {self.precio}             {self.total}')
            y=y-10
            if opcion=='TR':
                break


        print('      TOTAL:      ',self.totalpagar)
        print(self.separacion)
        c.drawString(210,530,f'      TOTAL:      {self.totalpagar}')
        c.drawString(210,520,self.separacion)


        self.dinerorecibido=int(input(f'EL COSTO A CUBRIR ES DE {self.totalpagar} '))
        self.cambio=self.dinerorecibido-self.totalpagar
        print('              FORMA DE PAGO','\n     EFECTIVO       ',self.dinerorecibido,'\n     CAMBIO       ',self.cambio)
        print(self.separacion)
        c.drawString(210,510,'                       FORMA DE PAGO')
        c.drawString(210,500,f'       EFECTIVO       {self.dinerorecibido}')
        c.drawString(210,490,f'       CAMBIO       {self.cambio}')
        c.drawString(210,480,self.separacion)


        print('         ¡GRACIAS POR SU COMPRA!','\n¡COMPRA UN DISPOCITIVO Y UNETE A NUESTRA FAMILIA!')
        c.drawString(210,470,'                  ¡GRACIAS POR SU COMPRA!')
        c.drawString(210,450,'¡COMPRA UN DISPOCITIVOY UNETE A NOSOTROS!')
        c.drawImage('Untitled.png',300,h-405,width=60,height=60)
        c.save()

passw=str(input('ingrese la clave'))

while passw !='2426':
    passw=str(input('contraseña incorrecta, ingrese la clave nuevamente '))

if passw=='2426':

    cliente=factura(str(input('nombre del cliente')),str(input(' documento del cliente')))
    cliente.factura_com()
Image('Untitled.png',width=100,height=100)
```
